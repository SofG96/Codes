# Query description

**Set the initial request in a specific language and time zone.**
**If necessary, set also the eventual proxies.**


*   *Language*: `hl = 'en-US'`, `hl = 'en-GB'`, `hl = 'FR'`, `hl = 'IT'`. 
*   *Time zone*: `tz=360` for CST, `tz=60` for CET

Example:

```
pytrends = TrendReq(hl='en-US', 
                    tz=60, 
                    timeout=(10,25), 
                    proxies=['https://34.203.233.13:80','https://35.201.123.31:880'], 
                    retries=2, 
                    backoff_factor=0.1, 
                    requests_args={'verify':False})
```


````
pytrends = TrendReq(hl='en-US', 
                    tz=60)
````

**Define your query**

The query is defined by various arguments:

1.   `kw_list` (required) is the list of countries of your choosing. It contains up to five terms.
      
      Example: `kw_list = ['DE', 'FR']`
2.   `cat` it is a number that defines the category of the query (e.g. Risk Management, Financial Markets, Fiscal Policy News). It can be useful to narrow down the search to one topic.

     More information about all available categories [here](https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories). 
     
     Example: `cat = 1240` stands for company earnings.

1.   `timeframe` is the time period. It can be:
      
      - `'all'`

      - specific dates in format DDDD-MM-YY (e.g. `'2020-02-19 2020-12-01'`)

      - specific dates and hours in format DDDD-MM-YYTHH (e.g. `'2020-03-12T14 2020-03-12T16'`)

      - relative hourly period: `'now 1-H'` or `'now 4-H'`

      - relative daily period: `'now 1-d'` or `'now-7d'`

      - relative monthly period: `'today 1-m'`, `'today 3-m'`, `'today 12-m'`

      - relative yearly period: `'today 5-y'`

    *Pay attention*: the time component is based on the UTC time zone.
      
2.   `gprop` is the Google property on which the search should be filtered.

      - `'images'` for Google Images

      - `'news'` for Google News

      - `'froogle'` for Google Shopping

      - `'youtube'` for Youtube

#Temporal and geographical search for a list of words and a set of countries

````
# Set your query in a specific language and time zone (CET is 60, CST is 360)
pytrend = TrendReq(hl='en-US',
                   tz=60) 

# Define a list of words of your interest
kw_list = ['covid','coronavirus','icu']

#Define the countries of interest
country = ['DE', 'FR', 'IT', 'ES']

# Run your query (in loop)
for i in country:
  pytrend.build_payload(kw_list,
                        cat=0,
                        timeframe='all',
                        geo=i,
                        gprop='')
  
  #Interest over time
  trend_time = pytrend.interest_over_time()
  print(trend_time)

  #Geographical heterogeneity
  trend_geo = pytrend.interest_by_region()
  print(trend_geo)

  #Plot the temporal results
  plot_data = trend_time.iloc[:,0:len(trend_time.columns)-1]
  plot_data = plot_data.reset_index() \
    .melt(id_vars='date', var_name='phrase', value_name='trend')
  
  time = sns.relplot(data=plot_data, x='date', y='trend',
                kind='line', row='phrase', height=1.5,
                aspect=6.7)
  
  #Plot the geographical results
  plot_data = trend_geo.iloc[:,0:len(trend_geo.columns)-1]
  plot_data = plot_data.reset_index() \
    .melt(id_vars='geoName', var_name='phrase', value_name='trend')

  geo = sns.relplot(data=plot_data, x='geoName', y='trend',
                kind='line', row='phrase', height=1.5,
                aspect=6.7)
  
  #Plot the results (nested loop)
  #for word in kw_list: 
    
    #trend_time.reset_index().plot(x='date', y=word, 
    #                              figsize=(60, 10), kind ='line')
    
    #trend_geo.reset_index().plot(x='geoName', y=word, 
    #                             figsize=(60, 10), kind ='bar')
````

**Filtering the geographical zone**

You can set the geographical zone via `resolution`.

It can be equal to `'COUNTRY'`, `'REGION'`, `'CITY'`, `'DMA'` (metropolitan region, only available for US).



````
pytrend = TrendReq(hl='en-US',
                   tz=60) 

pytrend.build_payload(kw_list = ['covid'],
                        cat=0,
                        timeframe='all',
                        geo='US',
                        gprop='')

trend_geo_country = pytrend.interest_by_region(resolution='DMA')
print(trend_geo_country)
````

**Single plot**

````
pytrend = TrendReq(hl='en-UK', tz=60)

kw_list = ['brexit']

pytrend.build_payload(kw_list,
                       cat=0,
                       timeframe='today 5-y',
                       geo='GB',
                       gprop='')

trend_df_single = pytrend.interest_over_time()
sns.lineplot(data=trend_df_single.iloc[:,0:1])
sns.despine()
````

**Time series decomposition**

````
df = trend_df_single.iloc[:,0:1]
decompose = sm.tsa.seasonal_decompose(df,
                                      model='additive')
````

````
sns.lineplot(data=decompose.trend)
sns.despine()
````
